```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# %matplotlib.inline

x = np.arange(1, 100)
y1 = np.log(x)
y2 = x ** 0.5
```

```python
plt.figure(figsize=(5,3), dpi=100)
plt.plot(x, y1, linestyle='-', color='orange')
plt.plot(x, y2, linestyle='--', color='red')
plt.title(r'$\alpha > \beta$')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.xticks(np.arange(min(x), max(x)+1, 10))
plt.grid(True)
plt.show()
```

```python
import matplotlib.gridspec as gridspec
# style
plt.style.use('ggplot')
plt.rcParams["axes.edgecolor"] = "0.15"
plt.rcParams["axes.linewidth"]  = 1

plt.suptitle("Figure Title", fontsize=20)
gs = gridspec.GridSpec(2, 2)

ax1 = plt.subplot(gs[0, 0])
ax1.plot(x, y1, linestyle=':', color='orange')
ax1.set_title("Plot 1")
ax1.set_xlabel('xlabel')
ax1.set_ylabel('ylabel')
plt.grid(True)

ax2 = plt.subplot(gs[0, 1])
ax2.plot(x, y2, linestyle='-.', color='blue')
ax2.set_title("Plot 2")
ax2.set_xlabel('xlabel')
ax2.set_ylabel('ylabel')
plt.grid(True)

ax3 = plt.subplot(gs[1, :])
ax3.plot(x, y1, linestyle='--', color='red')
plt.xticks(np.arange(min(x), max(x)+1, 10))
ax3.set_title("Plot 3")
ax3.set_xlabel('xlabel')
ax3.set_ylabel('ylabel')
plt.grid(True)

plt.tight_layout(rect=[0.1, 0.1, 0.9, 0.9]) # box in which the plots will be
plt.savefig("fig.pdf")
plt.show()
```
